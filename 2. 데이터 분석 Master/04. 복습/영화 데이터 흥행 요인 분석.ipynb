```python
movie = pd.read_csv("C:\\Users\\User\\Desktop\\패스트 캠퍼스 데이터 마스터\\파이썬\\Part3) 파이썬 데이터 분석 실습\\Data\\실습2_ 영화 데이터를 활용한 영화 흥행 요인 분석\\tmdb_5000_movies.csv")
movie
```

```python
credits = pd.read_csv("C:\\Users\\User\\Desktop\\패스트 캠퍼스 데이터 마스터\\파이썬\\Part3) 파이썬 데이터 분석 실습\\Data\\실습2_ 영화 데이터를 활용한 영화 흥행 요인 분석\\tmdb_5000_credits.csv")
credits.columns
```

```python
movies_df = movie[['id', 'budget', 'genres', 'title', 'release_date', 'revenue', 'vote_average', 'vote_count']]
credits_df = credits[['movie_id', 'crew', 'cast']]
```

## 데이터 결합 한 후 movie_id까지 있으면 id값이 두개가 있는거라 삭제해준다.

```python
data = pd.merge(movies_df, credits_df, left_on='id', right_on='movie_id').drop('movie_id', axis=1)
data.head()
```

#ROI 예산대비 수익익

```python
data['ROI'] = data['revenue'] / data['budget'] 
data.head()
```

# 감독 컬럼 만들기기

```python
data['crew'][0]
```

```python
import ast
```

```python
print(ast.literal_eval(data['crew'][0]))
```

```python
data['crew'] = data['crew'].apply(ast.literal_eval)
data.head()
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
```

```python
data['director'] = data['crew'].apply(get_director)
data.head()
```

# 배우 컬럼 만들기기

```python
data['cast'][0]
```

```python
data['cast_name'] = data['cast'].apply(lambda x : [i['name'] for i in ast.literal_eval(x)])
data.head()
```

# 장르컬럼 만들기기

```python
data['genres'] = data['genres'].apply(ast.literal_eval)
```

```python
def get_genres(x):
    if len(x) > 0:
        return x[0]['name']
```

```python
data['main_genre'] = data['genres'].apply(get_genres)
```

```python
data.head()
```

# 데이터 타입 변경

```python
data.info()
```

```python
data['release_date'] = pd.to_datetime(data['release_date'], format = '%Y-%m-%d')
data['id'] = data['id'].astype(str)
```

# 연도 월 컬럼 만들기

```python
data['year'] = data['release_date'].dt.year 
data['month'] = data['release_date'].dt.month 
```

# 결측지 제거

```python
data.dropna(inplace = True)
```

```python
data.info()
```

# 분석

```python
#[1] 연도별 흥행 수익은?
import matplotlib.pyplot as plt
```

```python
revenue_by_year = data.groupby('year')[['revenue']].sum().reset_index()
```

```python
plt.plot(revenue_by_year['year'], revenue_by_year['revenue'], color='r')
plt.grid(True)
plt.legend(loc='best')
plt.title('revenue_by_year')
plt.xlabel('year')
plt.ylabel('revenue')
plt.show()
```

## 가장 흥행한 영화 TOP10

```python
top = data.groupby('title')[['revenue']].sum().reset_index().sort_values('revenue', ascending=False).head(10)
plt.figure(figsize=(12, 8))
plt.bar(top['title'], top['revenue'],  color='blue' )   ## 폰트 사이즈
plt.legend(loc='best')
plt.title('revenue TOP 10')
plt.xlabel('title')
plt.xticks(rotation=-25, fontsize=10)
plt.ylabel('revenue')
plt.yticks(fontsize=12)
plt.show
```

## 예산 TOP 10, 투표수 TOP 10 영화는?

```python
for y in ['budget', 'vote_count']:
    top = data.groupby('title')[[y]].sum().reset_index().sort_values(y, ascending=False).head(10)
    plt.figure(figsize=(12, 8))  # plt.figure를 제거하면 차트가 하나로 나옴 
    plt.bar(top['title'], top[y],  color='blue')
    plt.show
```

```python
title_dic = {'budget' : '예산' , 'vote_count' : '투표수'}

for y in ['budget', 'vote_count']:
    top = data.groupby('title')[[y]].sum().reset_index().sort_values(y, ascending=False).head(10)
    plt.figure(figsize=(12, 8))
    plt.bar(top['title'], top[y],  color='blue')   ## 폰트 사이즈
    plt.title(f'{title_dic[y]} TOP 10 movie')
    plt.xlabel(y)
    plt.xticks(rotation=-25, fontsize=10)
    plt.ylabel(y)
    plt.yticks(fontsize=12)
    plt.show
```

## 연도별 장르별 수익

```python
# 제시 방법 ->  recenue_by_year_genre = data.query('year >= 1990').groupby(['year', 'main_genre'])[['revenue']].sum().reset_index()
recenue_by_year_genre = pd.pivot_table(data=data.query('year >= 1990'), index='year', columns='main_genre', values='revenue', aggfunc=sum)
```

```python
plt.figure(figsize=(12, 8))
recenue_by_year_genre.plot.bar(stacked=True, figsize=(10,7))
#plt.bar(recenue_by_year_genre['year'], recenue_by_year_genre['revenue'], color=recenue_by_year_genre['main_genre'])
plt.legend(loc='best')
plt.xticks(rotation=-25, fontsize=10)
plt.ylabel('revenue')
plt.show
```

```python
# 퍼센트로 변환
recenue_by_year_genre_pct = pd.pivot_table(data=data.query('year >= 1990'), index='year', columns='main_genre', values='revenue', aggfunc=sum, margins=True)
recenue_by_year_genre_pct = 100 * recenue_by_year_genre_pct.div(recenue_by_year_genre_pct.iloc[:,-1], axis = 0).drop('All').drop('All',axis=1)
# melt -> 행열 바꾸기
revenue_by_year_genre_pct = pd.melt(recenue_by_year_genre_pct.reset_index(), id_vars='year', value_name='pct').fillna(0)
revenue_by_year_genre_pct_1 = pd.pivot_table(data=revenue_by_year_genre_pct, index='year', columns='main_genre', values='pct', aggfunc=sum)
```

```python
plt.figure(figsize=(15, 10))
revenue_by_year_genre_pct_1.plot.bar(stacked=True, figsize=(10,7))
#plt.bar(recenue_by_year_genre['year'], recenue_by_year_genre['revenue'], color=recenue_by_year_genre['main_genre'])
plt.legend(loc='best')
plt.xticks(rotation=-25, fontsize=10)
plt.ylim([0,110])
plt.ylabel('revenue')
plt.show
```
