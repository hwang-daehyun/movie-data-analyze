```python
df = pd.read_csv("C:\\Users\\User\\Desktop\\패스트 캠퍼스 데이터 마스터\\파이썬\\Part3) 파이썬 데이터 분석 실습\\Data\\실습3_ 유통 데이터를 활용한 리텐션과 RFM 분석\\ecommerce_data.csv")
df.head
```

```python
df.info()
```

```python
# 분석을 위해 dropna에서 CustomerID를 지정하여 그 변수에 null값이 있는 행을 삭제

df.dropna(subset='CustomerID',inplace=True)
df.info()
```

```python
# InvoiceDate -> 날짜 데이터로, CustomerID는 오브젝트로 변경
# astype를 두번한 이유는 바로 str로 변경이 .0이 있는 상태로 문자형이 되어 int로 한번 수정후 사용용

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format = '%m/%d/%Y %H:%M')
df['CustomerID'] = df['CustomerID'].astype(int).astype(str)
df.info()
```

```python
# InvoiceDate 날짜와 연도 분리

df['date_ymd'] = df['InvoiceDate'].dt.date.astype('datetime64[ns]')
df['year'] = df['InvoiceDate'].dt.year
df.info()
```

```python
# 매출 컬럼 추가하기

df['amount'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
# Quantity에서 마이너스는 취소건이라 해당 데이터를 제외한 데이터만 살리기

data=df[df['Quantity'] > 0]
#df.query('Quantity' > 0) -> 또다른 방법법

data
```
