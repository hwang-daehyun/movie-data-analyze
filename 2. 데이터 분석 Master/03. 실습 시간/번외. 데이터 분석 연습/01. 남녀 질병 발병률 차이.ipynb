```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

## 데이터 불러오기

```python
df = pd.read_csv("C:\\Users\\User\\Desktop\\한국산업안전보건공단_성별 질병재해자수_20231231.csv", encoding = 'cp949')
df
type(df)
```

# 데이터를 세워서 보기

```python
data = df.transpose().reset_index()
type(data)
```

# index와 0 1 을 바꾸고 제목말고 데이터만 불러오기

```python
data.columns = ["구분", "남자", "여자"]
data = data.iloc[1:]
type(data)
```

# 구분 데이터에서 "질병재해자수" 삭제하기

```python
data["연도"] = data["구분"].str.split("년").str[0]
```

# 연도 컬럼이 있으니 구분 컬럼은 삭제

```python
data.drop(columns='구분', inplace=True)
```

```python
data['연도'] = pd.to_datetime(data['연도'], format='%Y')
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['연도'], data['남자'], label= 'male', color='blue')
plt.plot(data['연도'], data['여자'], label= 'female', color='red')
plt.legend(loc = 'upper left')
plt.xlabel('Year')
plt.ylabel('Sex')
plt.show()
```
