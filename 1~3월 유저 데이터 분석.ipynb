```python
day_total = pd.read_csv("C:\\Users\\dhwhang\\Downloads\\20250408_조사참여리스트_total.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
day_total.columns
```

```python
day_total['join_datetime'] = pd.to_datetime(day_total['join_datetime'], format='%Y-%m-%d %H:%M:%S')
```

```python
day_total['click_time'] = day_total['join_datetime'].dt.hour
```

```python
day_total['click_time']
```

```python
# 시간별 참여 빈도
click_total = day_total.groupby('click_time').count().reset_index()


click_total
```

```python
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'


plt.figure(figsize=(20,8))
plt.plot(click_total['click_time'], click_total['panel_seq'], marker = 'o')
plt.grid()
plt.xlabel('참여 시간', fontsize = 15)
plt.ylabel('빈도', fontsize = 15)
plt.xlim([0,23])
plt.xticks(np.arange(0,24))
plt.yticks(np.arange(0,100001,10000))
plt.show()
```

```python
# 회사에 따른 시간별 참여 빈도

day_brand_group = day_total.groupby(['click_time','company_num'])['panel_seq'].count().reset_index()
```

```python
pivot_company = day_brand_group.pivot(index='click_time', columns='company_num', values='panel_seq')
```

```python
company_value = {3:'마켓링크',4:'트렌드엠알'}

pivot_df = pivot_company.rename(columns=company_value)


pivot_df.plot(figsize=(12,6), marker='o')
plt.grid()
plt.legend(fontsize = 16)
plt.title('패널사별 참여 시간대')
plt.xlabel('시간', fontsize = 15)
plt.ylabel('빈도', fontsize = 15)
plt.xlim([0,23])
plt.xticks(np.arange(24))
plt.yticks(np.arange(0,100001,10000))
```

```python
df_job = pd.read_excel("D:\\프로파일\\마켓링크\\마켓링크(250409)\\250407_2025 서베이링크 패널 프로파일 조사_(JOB).xlsx")
```

```python
df_job.head()
```

```python
# 프로파일 데이터와 참여 데이터 합치기

df_total_JOB = pd.merge(day_total,df_job, left_on= "panel_seq",right_on='UID')
```

```python
df_total_JOB.head()
```

```python
# 필요한 변수만 가져오기

day_total_job = df_total_JOB[["company_num","panel_seq","campaign_id","click_time","LOI","gender","age","CITY_NAME","AQ9"]]
```

```python
day_total_job.head()
```

```python
# Nan 확인

day_total_job.info()
```

```python
def age(age):
    if age == 1:
        return 1
    elif age == 2:
        return 2
    elif age in [3,4,5,17,0]:
        return 3
    else:
        return 4
```

```python
day_total_job["JOB"] = day_total_job["AQ9"].apply(age)
 
```

```python
day_total_job.head()
```

```python
# 직업에 따른 시간대별 빈도

day_job_group= day_total_job.groupby(['click_time','JOB'])['panel_seq'].count().reset_index()
```

```python
pivot_df = day_job_group.pivot(index='click_time', columns='JOB', values='panel_seq')
```

```python


# 음수 기호가 깨지지 않도록 설정
# plt.rcParams['axes.unicode_minus'] = False

label_value = {1:'학생(중,고,대)',2:'주부',3:'기타(무직 및 기타)',4:'직장인'}

pivot_df = pivot_df.rename(columns=label_value)


pivot_df.plot(figsize=(12,6), marker='o')
plt.grid()
plt.legend(fontsize = 16)
plt.title('직업에 따른 참여 시간대')
plt.xlabel('시간', fontsize = 15)
plt.ylabel('빈도', fontsize = 15)
plt.xlim([0,23])
plt.xticks(np.arange(0,24))
plt.yticks(np.arange(0,100001,10000))
```

```python
# 성별에 따른 조사 참여 빈도 및 시간대 빈도

day_sex_group = day_total_job.groupby(['click_time','gender'])['panel_seq'].count().reset_index()
```

```python

day_sex_pivot = day_sex_group.pivot(index='click_time', columns='gender', values='panel_seq')
```

```python
sex_label_value = {1:'여자',2:'남자'}

pivot_sex = day_sex_pivot.rename(columns=sex_label_value)
sex_color=['red','blue']


pivot_sex.plot(figsize=(12,6), marker='o', color=sex_color)
plt.grid()
plt.legend(fontsize = 16)
plt.title('성별 따른 참여 시간대')
plt.xlabel('시간', fontsize = 15)
plt.ylabel('빈도', fontsize = 15)
plt.xlim([0,23])
plt.xticks(np.arange(0,24))
plt.yticks(np.arange(0,100001,10000))
```
